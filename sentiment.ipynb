Passo 2: Importando as bibliotecas para análise de
sentimento

```python
import spacy
from spacytextblob.spacytextblob import SpacyTextBlob
```

Passo 3: Definindo o modelo e a pipeline a serem
utilizadas na análise

```python
nlp = spacy.load('en_core_web_sm')
nlp.add_pipe('spacytextblob')
```

Passo 4: Definindo o texto inicial a ser analisado para
verificação/validação da biblioteca.

```python
user_input = ' This is a wonderful campsite. I loved the serenity and the birds chirping in the morning. '
doc = nlp(user_input)
```

Passo 5: Calcular e exibir a pontuação de sentimento do texto inicial (user_input), que varia entre -1 (negativo) e 1 (positivo).

```python
input_polarity  =  doc._.blob.polarity
sentiment = {
    'score': input_polarity
}

print(sentiment)
```
